```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
wrs=[]
aes=[]
for a2 in np.linspace(1.1,3.5,100):
    a1=1.
    #a2=2.8
    aes+=[a2]
    
    P1=1
    P2=np.sqrt(a2**3)

    w1=2*np.pi/P1
    w2=2*np.pi/P2

    Ps=1/(1/P1-1/P2)
    ws=2*np.pi/Ps
    #print("Synodic period and velocity:",Ps,ws)

    t=np.linspace(0,Ps,100)
    q1=w1*t
    q2=w2*t

    x1=a1*np.cos(q1)
    y1=a1*np.sin(q1)

    x2=a2*np.cos(q2)
    y2=a2*np.sin(q2)

    r=np.sqrt((x1-x2)**2+(y1-y2)**2)
    q=np.arctan2((y2-y1),(x2-x1))*180/np.pi
    w=(q[1:]-q[:-1])/(t[1:]-t[:-1])*np.pi/180

    wrs+=[w[0]/w2]
```

```python
fig,axs=plt.subplots(2,1)

axs[0].plot(t,r)
axs[1].plot(t[:-1],w)
#axs[1].set_axhline(-ws)
axs[1].axhline(+ws,color='r')
#axs[1].set_ylim((-5,5))
```

```python
fig,axs=plt.subplots(1,1)
axs.plot(aes,wrs)
```
